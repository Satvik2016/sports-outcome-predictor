```python
from nba_api.stats.endpoints import leaguegamefinder
import pandas as pd
import time

all_games = []

for season_year in range(2012, 2024):
    season_str = f"{season_year}-{str(season_year+1)[-2:]}"
    print(f"Pulling {season_str}...")
    try:
        gamefinder = leaguegamefinder.LeagueGameFinder(
            season_nullable=season_str,
            league_id_nullable='00'
        )
        games = gamefinder.get_data_frames()[0]
        all_games.append(games)
        time.sleep(1) 
    except Exception as e:
        print(f"  Skipped {season_str}: {e}")
        continue

df = pd.concat(all_games, ignore_index=True)


df = df.rename(columns={
    'TEAM_ABBREVIATION': 'team',
    'GAME_DATE':         'date',
    'MATCHUP':           'matchup',
    'WL':                'result_str',
    'PTS':               'team_score',
})


df['result'] = (df['result_str'] == 'W').astype(int)


df['is_home'] = df['matchup'].apply(lambda x: 0 if '@' in x else 1)
df['opponent'] = df['matchup'].apply(lambda x: x.split()[-1])


df.to_csv('data/raw/nba_games.csv', index=False)
print(f"Done! Saved {len(df)} rows.")
print(df[['team','date','opponent','is_home','team_score','result']].head(10))
```
